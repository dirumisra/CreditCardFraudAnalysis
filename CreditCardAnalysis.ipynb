**Splitting the data into train & test data**

```python
y = ccdf["Class"]
X = ccdf.drop('Class', axis = 1)
```

**Feature Scaling using Standard Scaler**

*Feature Scaling is a technique to standardize the independent features present in the data in a fixed range. It is performed during the data pre-processing to handle highly varying magnitudes or values or units*

**We need to scale Amount column.**

```python
# As PCA is already performed on the dataset from V1 to V28 features, we are scaling only Amount field
scaler = StandardScaler()

#Scaling the train Data
X_train[["Amount"]]= scaler.fit_transform(X_train[["Amount"]])

# Transformation the test data
X_test[["Amount"]] = scaler.transform(X_test[["Amount"]])
```

**let's Plot the distribution of a variable to handle skewness**

Skewness is a way to figure out if the stuff you're looking at is balanced or not. Imagine you have a bunch of data, like people's ages. If most of the ages are kind of in the middle, and there aren't too many really young or really old people, that's balanced or symmetrical. But if you have a lot of young people and only a few old people, or the other way around, that's not balanced or asymmetrical.

Skewness helps us put a number on how lopsided or balanced the data is. It's important because it tells us about the shape of the data. For example, if you're looking at a group of test scores, knowing if they're mostly bunched up in the middle or spread out all over can be helpful. Skewness gives us a clue about this, and it's handy when we're trying to understand the main values and how spread out the data is.

To calculate skewness, statisticians often use mathematical formulas that involve the mean, median, and standard deviation of the data. Positive skewness indicates a longer right tail, while negative skewness indicates a longer left tail. A skewness value of 0 indicates a perfectly symmetrical distribution.
