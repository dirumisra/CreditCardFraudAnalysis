```python
## Observe the distribution of our classes

classes = ccdf['Class'].value_counts()
normalshare = round(classes[0]/ccdf['Class'].count()*100,2)
fraudshare = round(classes[1]/ccdf['Class'].count()*100,2)

normalshare,fraudshare
```

#### Create a bar plot for the number and percentage of fraudulent vs non-fraudulent transcations

```python
plt.figure(figsize = (10,5))

plt.subplot(1,2,1)
sns.countplot(ccdf["Class"])
plt.ylabel("Number of Transaction")
plt.xlabel("Class")
plt.title("Credit Card Fraud Class - data unbalance")

plt.subplot(1,2,2)
fraud_percentage = {'Class' : ['Non-Fraudulent', 'Fraudulent'],'Percentage' : [normalshare, fraudshare]}
ccdf_fraud_percentage = pd.DataFrame(fraud_percentage)
sns.barplot(x='Class',y = 'Percentage',data = ccdf_fraud_percentage)
plt.title('Percentage of Fraudulent vs Non-Fraudulent Transactions')
plt.show()
```
